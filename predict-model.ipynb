```python
import tensorflow as tf
import numpy as np
import cv2

def predict_skin_disease(img):
    model = tf.keras.models.load_model('/kaggle/input/h5-files/model.h5')
    vgg_model = tf.keras.models.load_model('/kaggle/input/h5-files/vgg_model.h5')
    img = cv2.imread(img)
    img = cv2.resize(img, (180,180))
    img = np.array(img) / 255.0
    img = np.expand_dims(img, axis=0)
    img = vgg_model.predict(img)
    img = img.reshape(1,-1)
    pred = model.predict(img)[0]
    predicted_class = np.argmax(pred)
    disease_list = ['Acne and Rosacea Photos',
            'Normal',
            'vitiligo',
            'Tinea Ringworm Candidiasis and other Fungal Infections',
            'Melanoma Skin Cancer Nevi and Moles',
            'Eczema Photos']
    return disease_list[predicted_class]
```
